```python
import itertools
import random

# sttike와 ball을 계산하는 함수
def judge(a, b):
    strike, ball = 0, 0
    for i in range(3):
        if a[i] == b[i]: strike += 1
        if a[i] in b: ball += 1
    return strike, ball-strike

# 추측을 할 수 있는 모든 경우를 guessList에 저장
t = [i for i in range(1, 10) ]
guessList = [k for k in itertools.permutations(t, 3) ]
# 램덤하게 추측하도록 미리 guessList를 섞는다.
random.shuffle(guessList)
round = 1
history = [ ]
for guess in guessList:
    isOk = True
    for h in history:
        s, b = judge(h[0], guess)
        if s != h[1] or b != h[2]:
             isOk = False
             break
    if not isOk: continue
    # 추측한 숫자를 출력
    print("Guess :", "".join(map(str, guess)))
    s, b = map(int, input().split())
    if s == 3:
        print("Scroe : %d"%round)
        break
    history += [ (guess, s, b) ]
    round += 1
```

```python
# lab6 - Sudoku / 플래그, 반복시 if 여러개 X - 미리 만들어 놓은 것 사용

# 입력을 받습니다.
lineFlag = [ [ False ]*10 for _ in range(9) ]
columnFlag = [ [ False ]*10 for _ in range(9) ]
blockFlag = [ [ [False]*10 for _ in range(3) ] for _ in range(3) ]

def solve(cur, blank, sudoku):
    global lineFlag
    global columnFlag
    global blockFlag
    if cur == len(blank): return True
    (r, c) = blank[cur]
    for k in range(1, 10):
        if lineFlag[r][k]: continue
        if columnFlag[c][k]: continue
        if blockFlag[r//3][c//3][k]: continue
        sudoku[r][c] = k
        lineFlag[r][k] = True
        columnFlag[c][k] = True
        blockFlag[r//3][c//3][k] = True
        if solve(cur+1, blank, sudoku): return True
        sudoku[r][c] = 0
        lineFlag[r][k] = False
        columnFlag[c][k] = False
        blockFlag[r//3][c//3][k] = False
    return False

# 입력을 받습니다.
sudoku = [ ]
for _ in range(9):
    sudoku.append(list(map(int, input().split())))

blank = [ ]
for r in range(9):
    for c in range(9):
        v = sudoku[r][c]
        if v == 0:
            blank.append((r, c))
        else:
            lineFlag[r][v] = True
            columnFlag[c][v] = True
            blockFlag[r//3][c//3][v] = True

solve(0, blank, sudoku)
for r in range(9):
    print(*sudoku[r])
```
